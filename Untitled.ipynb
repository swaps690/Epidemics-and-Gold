```python
index = pd.read_csv('data_ebola.csv',index_col='Date',parse_dates=True)
```

```python
index.plot(figsize=(12,8))
```

```python
df = pd.DataFrame()
```

```python
df['gg']=gg['Last Price']
```

```python
df['cc']=cc['Last Price']
```

```python
df.head()
```

```python
df = df.iloc[::-1]
```

```python
df.head()
```

```python
df.drop(df.index[0])
```

```python
index.drop(index.index[-1])
```

```python
df['market']=index['Close']
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```

```python
X = df[["gg","cc"]] ## X usually means our input variables (or independent variables)
y = df["market"] ## Y usually means our output/dependent variable
X = sm.add_constant(X) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)
predictions = model.predict(X)

# Print out the statistics
model.summary()
```
